DadmaTools' normalizer:

```python
from dadmatools.normalizer import Normalizer

normalizer = Normalizer(
    full_cleaning=False,
    unify_chars=True,
    refine_punc_spacing=True,
    remove_extra_space=True,
    remove_puncs=False,
    remove_html=False,
    remove_stop_word=False,
    replace_email_with="<EMAIL>",
    replace_number_with=None,
    replace_url_with="",
    replace_mobile_number_with=None,
    replace_emoji_with=None,
    replace_home_number_with=None
)

text = """
<p>
دادماتولز اولین نسخش سال ۱۴۰۰ منتشر شده.
امیدواریم که این تولز بتونه کار با متن رو براتون شیرین‌تر و راحت‌تر کنه
لطفا با ایمیل dadmatools@dadmatech.ir با ما در ارتباط باشید
آدرس گیت‌هاب هم که خب معرف حضور مبارک هست:
 https://github.com/Dadmatech/DadmaTools
</p>
"""

print('output text when is normalized : ', normalizer.normalize(text))
```

```python
# full normalized version
normalizer = Normalizer(full_cleaning = True)
print('output text when using full_cleaning normalizer is: ', normalizer.normalize(text))
```

NLP models can be used via pipeline:

```python
# making formal:
from dadmatools.pipeline.informal2formal.main import Informal2Formal
translator = Informal2Formal()

print(translator.translate('اینو اگه خواستین میتونین واسه خودتون تست کنین '))
```

```python
# tokenizing:
import dadmatools.pipeline.language as language

# as tokenizer is the default tool, it will be loaded even without calling
pips = 'lem,pos,ner,dep,spellchecker,kasreh,sent,itf'
nlp = language.Pipeline(pips)
```

```python
text = 'من صادق جعفری‌زاده به عنوان توسعه‌دهنده دادماتولز از شرکت دادماتک هستم. من به لوزامبورگ خواهم رفت.'
doc = nlp(text)
doc
```

#1-SpellCheker

```python
text = 'غزا را برای بغیه آماد کدره و روی میز بگزار'
doc = nlp(text)
doc['spellchecker']
```

همان طور که میبینیم، غزا به غذا و کدره به کرده درست شده است ولی:

1.    به‌جای تبدیل بغیه به بقیه، آن را به بسته تبدیل کرد

2.   به جای تبدیل آماد به آماده، آن را تغییری نداد
3.   به‌جای تبدیل بگزار به بگذار، آن را به برگزار تبدیل کرد



#2-ITF

```python
text = 'وقتی خواندیش با من راجع بهش حرف بزن'
doc = nlp(text)
doc['itf']
```

 همانطور که مشخص است خواندیش رسمی نیست و باید از آن را خواندی استفاده شود

#3-LEM


```python
text = 'من گربه‌ای را میدیدم که داشت غذا میخورد'
doc = nlp(text)
# doc['lemma'] considering the example
for sentence in doc['sentences']:
  for token in sentence['tokens']:
    print(token['lemma'])
```

 همانطور که مشخص است گربه نباید ریشه یابی متفاوتی داشته باشد،
  همچنین ریشه "دیدن" می تواند بن مضارع "بین" باشد و "آ" برای آن اشتباه است

#4-POS

```python
text = 'ستاره آب را نوشید و به خانه‌اش رفت'
doc = nlp(text)
# doc['pos'] considering the example
print(f"Text    upos    xpos")
for sentence in doc['sentences']:
  for token in sentence['tokens']:
        print(f"{token['text']}     {token['upos']}     {token['xpos']}")
```

 آب به عنوان مفعول می باشد اما فعل در نظر گرفته شده است
همچنین ترکیب "به خانه‌اش رفت" فعل نیست و "به" حرف اضافه و "خانه‌اش" متمم است و اسم هستند

#5-NER


```python
text = ' ستاره باباجانی با پرواز ایرباس  برای شرکت در مراسم هالووین رفت'
doc = nlp(text)
# doc['ner'] considering the example
print(f"Text     NER")
for sentence in doc['sentences']:
  for token in sentence['tokens']:
        print(f"{token['text']}     {token['ner']}")
```

همان طور که میبینیم تگ‌های مربوط به "ستاره باباجانی" اشتباه هستند و باید بعنوان اسم شناخته میشد. همچنین "ایرباس" و "هالووین" ارگانیزیشن هستند.

#6-DEP

```python
text = 'ستاره آب را نوشید و به خانه‌اش رفت '
doc = nlp(text)
print(f"Text    Head    Deprel")
for sentence in doc['sentences']:
  dependency_parsing = sentence['tokens']
  for token in dependency_parsing:
    print(token['text'],'\t',token['head'],'\t',token['deprel'])
```


  مشخص شده است amod باشد اما obj همانطور که مشخص است نقش آب باید
  همچنین کلمه آب زیر نظر کلمه هشتم قرار گرفته که باید زیر نظر "نوشید" قرار بگیرد
  و همینطور اشتباهات دیگر که قابل مشاهده‌اند
